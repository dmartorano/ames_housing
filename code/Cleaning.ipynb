## Imports

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../datasets/train.csv")
```

## Checking and Cleaning

```python
train.head()
```

```python
train.columns = train.columns.str.lower()
train.columns = train.columns.str.replace(" ", "_")
```

```python
pd.options.display.max_columns = 81 
# Expands printed dataframes so all base rows are visible
```

```python
train.head(20)
```

### Defining and Assessing Vital Features

The vital features in this dataset for addressing the problem statement are the neighborhood of the home and any feature that pertains directly to the strength, condition, and quality of the basement of the home.

```python
train["bsmt_half_bath"].fillna(0, inplace = True)
# Assumes that NaN values mean that there is no half bath in the basement
```

```python
train["bsmt_full_bath"].fillna(0, inplace = True)
# Assumes that NaN values mean that there is no full bath in the basement
```

```python
train = train[train["total_bsmt_sf"].notnull()]
# Makes dataframe not include any rows where total basement square footage is a null value, 
# found on https://stackoverflow.com/questions/13413590/how-to-drop-rows-of-pandas-dataframe-whose-value-in-a-certain-column-is-nan
```

```python
train = pd.get_dummies(train, columns=["neighborhood"], dtype = int)
# Crucial to determining proximity to ISU
```

```python
train = pd.get_dummies(train, columns=["bsmt_qual"], dtype = int)
```

```python
train = pd.get_dummies(train, columns=["bsmt_cond"], dtype = int)
```

```python
train = pd.get_dummies(train, columns=["bsmt_exposure"], dtype = int)
```

```python
train = pd.get_dummies(train, columns=["bsmtfin_type_1"], dtype = int)
```

```python
train = pd.get_dummies(train, columns=["bsmtfin_type_2"], dtype = int)
```

```python
pd.options.display.min_rows = 50
# Makes it possible to see all rows with null values in .isnull method
```

```python
train.isnull().sum().sort_values(ascending=False)
# All features vital to DS problem ensured to not include null values
```
